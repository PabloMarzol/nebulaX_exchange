```python
from alpaca.data.historical import StockHistoricalDataClient
from alpaca.data.requests import StockBarsRequest
from alpaca.data.timeframe import TimeFrame

from alpaca.trading.client import TradingClient
from alpaca.trading.requests import GetAssetsRequest
from alpaca.trading.enums import AssetClass

from dotenv import load_dotenv
from datetime import datetime

import os
import polars as pl

load_dotenv()
```

```python
data_client = StockHistoricalDataClient(os.getenv("ALPACA_API_KEY"), os.getenv("ALPACA_API_SECRET"))
trading_client = TradingClient(os.getenv("ALPACA_API_KEY"), os.getenv("ALPACA_API_SECRET"))
```

```python
# Creating request object
req_params = StockBarsRequest(
    symbol_or_symbols=["AAPL"],
    timeframe=TimeFrame.Day,
    start=datetime(2022, 1, 1),
    end=datetime(2022, 6, 1)   
)
stocksData = data_client.get_stock_bars(req_params)

tickers = stocksData["AAPL"]

# ticker_data = [ticker.__dict__ for ticker in stocksData]

df = pl.DataFrame(tickers)
df
```

```python
search_params  = GetAssetsRequest(asset_class = AssetClass.US_EQUITY)
tickers = trading_client.get_all_assets(search_params)

df = pl.DataFrame(tickers)
df.filter(pl.col("status") == "active")
```
